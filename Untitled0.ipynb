```python
########### VECTORIZATION TECHNIC #####
import gensim
y = data['class'].values
x = [d.split() for d in data['text'].tolist()]
type(x[0])
print(x[0])
```

```python
DIM = 100
w2v_model = gensim.models.Word2Vec(sentences=x, size=DIM, window=10, min_count=1)
```

```python
len(w2v_model.wv.vocab)
```

```python
#w2v_model.wv['love']
w2v_model.wv['india']
#w2v_model.wv.most_similar('gandhi')
#w2v_model.wv.most_similar('modi')
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(x)
```

```python
x = tokenizer.texts_to_sequences(x)
```

```python
tokenizer.word_index
```

```python
plt.hist([len(x) for x in x] , bins = 700)
plt.show()
```

```python
nos = np.array([len(x) for x in x])
len(nos[nos>1000])
```

```python
maxlen = 1000
x = pad_sequences(x, maxlen=maxlen)
```

```python
len(x[101])
```

```python
vocab_size = len(tokenizer.word_index) + 1
vocab = tokenizer.word_index
```

```python
def get_weight_matrix(model):
   weight_matrix = np.zeros((vocab_size , DIM))
   for word , i  in vocab.items():
      weight_matrix[i] = model.wv[word]
   return weight_matrix
```

```python
embedding_vectors = get_weight_matrix(w2v_model)
```

```python
embedding_vectors.shape
```

```python
#from sklearn.svm import LinearSVC
```

```python
model = Sequential()
model.add(Embedding(vocab_size, output_dim=DIM, weights = [embedding_vectors], input_length=maxlen , trainable= False))
model.add(LSTM(units=128))
#model.add(LinearSVC())
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
model.summary()
```

```python

x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
model.fit(x_train, y_train, validation_split=0.6, epochs=6)
```

```python
y_pred = (model.predict(x_test) >=0.7).astype(int)
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
print("LSTM Accuracy Score -> ",accuracy_score(y_test, y_pred))
print("\nConfusion Matrix of LSTM Classifier:\n")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report of LSTM Classifier:\n")
print(classification_report(y_test, y_pred))

#0.5 : 50%
```
